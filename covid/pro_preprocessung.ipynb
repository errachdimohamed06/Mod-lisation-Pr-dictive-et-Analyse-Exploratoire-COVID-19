```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('covid.csv', index_col=0, encoding = "ISO-8859-1")
data.head()
```

<h2><font color="green">PRE-PROCESSING</font></h2>


```python
df = data.copy()
df.head()
```

<h2><font color="green">Création des sous-ensembles (suite au EDA)</font></h2>


```python
missing_rate = df.isna().sum()/df.shape[0]
missing_rate
```

```python
blood_columns = list(df.columns[(missing_rate < 0.9) & (missing_rate >0.88)])
viral_columns = list(df.columns[(missing_rate < 0.80) & (missing_rate > 0.75)])
```

```python
key_columns = ['Patient age quantile', 'SARS-Cov-2 exam result']
df = df[key_columns + blood_columns + viral_columns]
df.head()
```

<h2><font color="green">
TrainTest - Nettoyage - Encodage</font></h2>


```python
from sklearn.model_selection import train_test_split
trainset, testset = train_test_split(df, test_size=0.2, random_state=0)
```

```python
def encodage(df):
    code = {'negative':0,
            'positive':1,
            'not_detected':0,
            'detected':1}
    
    for col in df.select_dtypes('object').columns:
        df.loc[:,col] = df[col].map(code)
        
    return df
```

```python
def imputation(df):
    #df['is na'] = (df['Parainfluenza 3'].isna()) | (df['Leukocytes'].isna())
    #df = df.fillna(-999) #renplir les données manquantes par la valeur -999
    df = df.dropna(axis=0)
    return  df    
```

```python
def feature_engineering(df):
    df['est malade'] = df[viral_columns].sum(axis=1) >= 1
    df = df.drop(viral_columns, axis=1)
    return df
```

```python
def preprocessing(df):    
    df = encodage(df)
    df = feature_engineering(df)
    df = imputation(df)
    
    X = df.drop('SARS-Cov-2 exam result', axis=1)
    y = df['SARS-Cov-2 exam result'].astype(int)  # Forcer l'encodage en entier
    
    print(y.value_counts())
    
    return X, y
```

```python
X_train,y_train=preprocessing(trainset)
X_train.shape
```

```python
X_test,y_test=preprocessing(testset)
X_test.shape
```

<h2><font color="green">
Modellisation</font></h2>


```python
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.model_selection import learning_curve
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import PolynomialFeatures
from sklearn.decomposition import PCA
```

<h2><font color="green">

Procédure d'évaluation</font></h2>


```python
def evaluation(model):
    
    model.fit(X_train, y_train)
    ypred = model.predict(X_test)
    
    print(confusion_matrix(y_test, ypred))
    print(classification_report(y_test, ypred))
    N, train_score, val_score = learning_curve(model, X_train, y_train,
                                              cv=4, scoring='f1',
                                               train_sizes=np.linspace(0.1, 1, 10))

    plt.figure(figsize=(12, 8))
    plt.plot(N, train_score.mean(axis=1), label='train score')
    plt.plot(N, val_score.mean(axis=1), label='validation score')
    plt.legend()
```

```python
model_1 = DecisionTreeClassifier(random_state=0)
evaluation(model_1)
```

```python
pd.DataFrame(model_1.feature_importances_, index=X_train.columns).plot.bar(figsize=(12, 8))
```

<h2><font color="green">l'importance des colonnes(Hematocrit,Leukocytes,est_malade) de notre dataset dans l'evalution de notre model (model_1)</font></h2>


```python
model_2 = RandomForestClassifier(random_state=0)
evaluation(model_2)
```

```python
pd.DataFrame(model_2.feature_importances_, index=X_train.columns).plot.bar(figsize=(12, 8))
```

<h2><font color="green">l'importance des colonnes(Platelets,Leukocytes,Monocytes,est_malade) de notre dataset dans l'evalution de notre model (model_2)</font></h2>


```python
pip_model_2 = make_pipeline(PolynomialFeatures(2), SelectKBest(f_classif, k=10),
                      RandomForestClassifier(random_state=0))
evaluation(pip_model_2)
```
